```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the PB Positive PB collisions data set
data = pd.read_csv("Plomo_pos")  #CSV

# Select the relevant columns (TOF and Moment)
X = data[['TPC', 'Momento']]

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-Means
kmeans = KMeans(n_clusters=5)  # Especifica el nmero de clusters deseado
kmeans.fit(X_scaled)

# Cluster labels
labels = kmeans.labels_

# Add cluster labels to the original dataframe
data['Cluster'] = labels

# Select the relevant columns for the new CSV file
output_data = data[['Event', 'Particula', 'Px', 'Py', 'Pz', 'Carga', 'Cluster']]

# Export the new dataframe to a CSV file
output_data.to_csv("PB-Pos-C1.csv", index=False)

# View the results
plt.scatter(X['Momento'], X['TPC'], c=labels, cmap='viridis')
plt.xlabel('TPC')
plt.ylabel('Momento')
plt.title('Clustering de Partculas con K-Means')
plt.colorbar(label='Cluster')

# Set limits on the x and y axes
plt.ylim(0, 700)
plt.xlim(0, 5)
plt.show()
```
